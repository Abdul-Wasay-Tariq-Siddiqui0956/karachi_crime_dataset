```python
import matplotlib.pyplot as plt
```

```python
town_counts = data['TOWN'].value_counts()

plt.figure(figsize=(12,6))
town_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Crimes per Town')
plt.xlabel('Town')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

```python
zone_counts = data['RISK_ZONE'].value_counts()

plt.figure(figsize=(4,4))
zone_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['red','orange','yellow','green','white'])
plt.title('Distribution of Risk Zones')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(data['LONGITUDE'], data['LATITUDE'], c=data['SEVERITY_SCORE'], cmap='Reds', alpha=0.6)
plt.colorbar(label='Severity Score')
plt.title('Crime Locations in Karachi')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
